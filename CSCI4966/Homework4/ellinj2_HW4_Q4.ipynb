#### Plot Images

```python
def plot_images(imgs1,imgs2,size, n = 10):
    plt.figure(figsize=(20, 4))
    for i in range(n):
        # display original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(imgs1[i].reshape(*size))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        # display reconstruction
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(imgs2[i].reshape(*size))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()
```

#### 4.2 (2) Load Cifar10 training and test data, convert to float32 and scale (0,1)

```python
# Your Code Here
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()

X_train.shape, X_test.shape
```

#### 4.3 (2) Autoencoders want a flat vector input

* Reshape data so each example is flattened,print the shapes

```python
# Your Code Here
X_train = X_train.reshape((len(X_train), np.prod(X_train.shape[1:])))
X_test = X_test.reshape((len(X_test), np.prod(X_test.shape[1:])))

X_train = X_train.astype("float32")/255.
X_test = X_test.astype("float32")/255.

X_train.shape, X_test.shape
```

#### 4.4 (2) Create Noisy input, use a noise factor of .3

```python
# Your Code Here
noise_factor = 0.3
X_train_noise = X_train + noise_factor * np.random.normal(size=X_train.shape)
X_test_noise = X_test + noise_factor * np.random.normal(size=X_test.shape)

X_train_noise = np.clip(X_train_noise, 0.0, 1.0)
X_test_noise = np.clip(X_test_noise, 0.0, 1.0)
```

#### 4.5 (2) Call plot_images with clean and noisy test images, and a tuple with the dimensions of the image

```python
plot_images(X_train, X_train_noise, (32,32,3))
plot_images(X_test, X_test_noise, (32,32,3))
```

#### 4.6 (10) Create Autoencoder with at least 3 encoding and decoding layers, output model summary

* Try different Hidden layer sizes

```python
# Your Code Here
input_dim = 3072
outer_dim = input_dim/6
middle_dim = outer_dim/2
inner_dim = middle_dim/2
output_dim = input_dim

input_img = Input(shape=(input_dim,))
encoder = Dense(outer_dim, activation="relu")(input_img)
encoder = Dense(middle_dim, activation="relu")(encoder)
encoder = Dense(inner_dim, activation="relu")(encoder)

decoder = Dense(middle_dim, activation="relu")(encoder)
decoder = Dense(outer_dim, activation="relu")(decoder)
decoded = Dense(output_dim, activation="sigmoid")(decoder)

model = Model(input_img, decoded)
model.summary()
```

#### 4.7 (2) Compile using Adam optimizer and binary_cross_entropy loss

```python
# Your Code Here
model.compile(optimizer="adam", loss="binary_crossentropy")
```

#### 4.8 (2) Fit the model with training data and validation data, use batch size of  256

```python
# Your Code Here
model.fit(X_train, X_train,
                epochs=20,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test))
```

#### 4.9 (2) Predict the test images, plot the  test images and the predicted images

```python
# Your Code Here
decoded_imgs = model.predict(X_test)
plot_images(X_test, decoded_imgs, (32,32,3))
```
